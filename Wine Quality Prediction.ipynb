Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

Data Collection

```python
# loading the dataset to a pandas dataframe
wine_dataset = pd.read_csv('/content/winequality-red.csv')
```

```python
# number of rows and columns in the dataset
wine_dataset.shape
```

```python
# First 5 rows of the dataset
wine_dataset.head()
```

```python
# checking for missing values
wine_dataset.isnull().sum()
```

Data Analysis and Visualization

```python
wine_dataset.describe()
```

Correlation

```python
correlation = wine_dataset.corr()
```

```python
plot = plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='pink')
```

Data Preprocessing

```python
# Seperate the data and level
X = wine_dataset.drop(columns='quality',axis=1)
Y = wine_dataset['quality']
```

```python
print(X)
```

```python
print(Y)
```

Label Binarization

```python
Y = wine_dataset['quality'].apply(lambda Y_value: 1 if Y_value>=7  else 0)
print(Y)
```

Train & Test split

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape , X_train.shape, X_test.shape)
```

```python
print(Y.shape ,Y_train.shape, Y_test.shape)
```

Training the model


Random Forest Classifier

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train,Y_train)
```

Model Evaluation

```python
# accuracy_score on training data
X_train_prediction = model.predict(X_train)
train_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy_score of the training data :",train_data_accuracy)
```

```python
# accuracy_score on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy_score of the test data:",test_data_accuracy)
```

Making a predictive System

```python
# taking input data
input_data = (8.5,0.28,0.56,1.8,0.092,35.0,103.0,0.9969,3.3,0.75,10.5)

# chaning the input data into a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print("Wine quality is bad")
else :
  print("Wine quality is good")


```
